```python
# while searching any key and if its not present we get error.
#To get value but if its not present we dont want to get error
# we can get

fruits.get('apple')
```

```python
fruits.items() #its under the list
```

```python
for i in fruits:
  print(i)
```

```python
for i in fruits:
  print(i, fruits[i])
```

```python
for i in fruits.keys():
  print(i)
```

```python
for i in fruits.items():
  print(i)
```

```python
for i in fruits.items():
  k,v = i
  print(k,v)
```

```python
for i in fruits.items():
  k,v = i
  print(k,v,sep=':')
```
